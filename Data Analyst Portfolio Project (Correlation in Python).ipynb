```python
#Importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# We need to read the source data to be used for the analysis

df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
# To check for missing data within the dataframe

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing)))
        
```

```python
# We need to know the types of data each colum is attributed with. We want to deal with budget and gross
print(df.dtypes)
```

```python
# We are dealing with empty rows in the dataset and also convert the float to int64
df = df.dropna() # This syntax deals with the empty rows in the dataset under each column
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
df
```

```python
# Creating a year column so we can use that for better correlation
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
# The syntax below gives a clear indication of the highest values in both columns to the lowest values.

df.sort_values(by=['gross'],inplace=False, ascending=False)
df.sort_values(by=['budget'],inplace=False, ascending=False)
```

```python
# Performing analysis on the dataset to drop any duplicates 

df['company'].drop_duplicates().sort_values(ascending=False) #Showing us what companies are distinct withint the dataframe.
```

 Now that our data has been checked and analyzed diligently, we will start looking at what columns we need to apply correlation to. 
 One needs to start asking questions as to what the dependent and independent variables in the correlation will be.
 
 # Quesitons.

```python
# Scatter plot of budget vs gross to see any correlation

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('budget vs gross earning')
plt.xlabel('gross earnings')
plt.ylabel('budget for film')
```

```python
# To verify from our graph, if they are correlated, we need to perform a regression plot on the graph. We will use 
# seaborn to perfrom the regression plot

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"black"}, line_kws={"color":"blue"})
```

```python
# Looking at the correlation invovled in the plotted data.
# There are different methods of Correlation, Pearson, kendall, Spearman

df.corr(method='pearson')
```

```python
# From the table above (correlation matrix), it could be seen that there's a high correlation between 'budget' and 'gross'

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show
```

```python
# The next category in our hypothesis is 'company'

# Company in this sense is not a number (they are objects) and so does other fields. So to convert that, we need to run a 
# for loop

df_number = df

for col_name in df_number.columns:
    if(df_number[col_name].dtype == 'object'):
        df_number[col_name] =  df_number[col_name].astype('category')
        df_number[col_name] =  df_number[col_name].cat.codes

df_number
```

```python
correlation_matrix = df_number.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show
```

```python
# Keeping the df_number in its correlation matrix to find an appropriate meaning to the values in the matrix

df_number.corr()
```

```python
# Let us try to unstack the matrix to see what we can find.
correlation_mat = df_number.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Syntax for sorting the pairs within the correlation matrix
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
# finding the high correlation with the sorted pairs
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
